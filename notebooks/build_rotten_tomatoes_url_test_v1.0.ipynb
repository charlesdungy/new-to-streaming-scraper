```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../data/processed/netflix_april.csv')
df.head()
```

```python
df_title = pd.DataFrame(df.Title)
df_title.head()
```

```python
df_title['search_url'] = df_title.Title.replace(to_replace=['\s+'], value='%20', regex=True)
df_title['search_url'] = 'https://www.rottentomatoes.com/api/private/v2.0/search/?limit=20&q=' + df_title['search_url']
print(df_title.to_string())
```
