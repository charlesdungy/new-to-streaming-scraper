```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json('../data/raw/rotten_tomatoes_april_scores.json')
```

```python
df.head()
```

```python
df[df.Score == ""]
```

```python
df.Score.replace("", "NR", regex=True, inplace=True)
```

```python
df['Year'] = df.Details.str.split(",").str[0]
df['Title_Category'] = df.Details.str.split(",").str[1]
df['Runtime'] = df.Details.str.split(",").str[2]

df.drop('Details', axis=1, inplace=True)
df.head()
```

```python
columns = ['Title', 'Year', 'Score', 'Title_Category', 'Runtime', 'URL']
df = df[columns]
df.head()
```

```python
df.duplicated(subset=['Title', 'Year', 'Title_Category', 'Runtime']).value_counts()
```

```python
df[df.duplicated(['Title', 'Year', 'Title_Category', 'Runtime'], keep=False)]
```

```python
df = df.drop_duplicates(subset=['Title', 'Year', 'Title_Category', 'Runtime'], keep='last')
```
