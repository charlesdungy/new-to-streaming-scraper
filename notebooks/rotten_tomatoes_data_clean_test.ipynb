```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json('../data/raw/rotten_tomates_movie_data.json')
```

```python
df.head()
```

```python
columns_no_actors = ['id', 'movie', 'tomato_score', 'release_date', 'runtime', 'url']
columns_actors = ['id', 'actors']
```

```python
df_no_actors = df[columns_no_actors]
df_no_actors.head()
```

```python
df_no_actors = df_no_actors.fillna(value = np.nan)
df_no_actors.runtime = df_no_actors.runtime.fillna('')
df_no_actors = df_no_actors.drop_duplicates(subset=['id'], keep='last')
df_no_actors.head()
```

```python
df_actors = df[columns_actors]
df_actors.head()
```

```python
df_actors = df_actors.explode('actors')
df_actors = df_actors.drop_duplicates(subset=['id', 'actors'], keep='last')
df_actors.head()
```
